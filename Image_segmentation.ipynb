# There are variours Image Segmentation techniques. Some of them are shown in this notebook


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# 1. Thresholding

```python
img=cv2.imread('elephant.png',cv2.IMREAD_UNCHANGED)

plt.imshow(img)
plt.title('Original Image')
plt.show()
```

```python
ret,img_th=cv2.threshold(img,127,255,cv2.THRESH_BINARY)

    
plt.imshow(img_th)
plt.title(' Binary Thresholding')

plt.show()
```

```python
img =cv2.imread('elephant.png',cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY ) 
img_th3 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

plt.imshow(img_th3)
plt.title(' Adaptive Gaussian Golobal Thresholding')

plt.show()
```

```python
img=cv2.imread('polo.png')
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY ) 
img_th4 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)



plt.imshow(img_th4)
plt.title(' Adaptive Mean Golobal Thresholding')

plt.show()
```

# 2. Edge based Segmentation

```python
img=cv2.imread('elephant.png',cv2.IMREAD_UNCHANGED)
edges = cv2.Canny(img,100,200)

plt.imshow(edges)
plt.title('Edge Image')
plt.show()
```
