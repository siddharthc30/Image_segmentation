# There are variours Image Segmentation techniques. Some of them are shown in this notebook


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# 1. Thresholding

```python
img=cv2.imread('elephant.png',cv2.IMREAD_UNCHANGED)

plt.imshow(img)
plt.title('Original Image')
plt.show()
```

```python
ret,img_th=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,img_th2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
    
plt.imshow(img_th)
plt.title(' Binary Thresholding')

plt.show()
```

```python

img=cv2.imread('elephant.png',cv2.IMREAD_GRAYSCALE)
img_th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

plt.imshow(img_th)
plt.title(' Adaptive Gaussian Golobal Thresholding')

plt.show()
```
